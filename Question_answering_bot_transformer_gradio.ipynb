```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering")
```

```python


def bot(Context,Question):
  QA_input = {'question' : Question , 
             'context' : Context }
  res = qa_pipeline(QA_input)
  return res['answer']

```

```python
import gradio
gradio.Interface(bot,['text','text'],'text').launch(share=True)
```
